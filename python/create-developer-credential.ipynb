### Create an API key

```python
from arcgis import __version__, GIS
from datetime import datetime, timedelta
```

```python
print(__version__)
```

```python
portal = GIS(profile="DTS2026")
```

```python
print(f"connected to {portal.properties.name}")
```

```python
developer_credential_manager = portal.admin.developer_credentials
```

```python
three_days = datetime.now() + timedelta(days=3)
```

```python
new_developer_credential = developer_credential_manager.create(
    title="Progammatically created developer credential in python",
    privileges=["premium:user:geocode", "premium:user:spatialanalysis","premium:user:demographics"],
    tags=["2026","developer","technology","summit"],
    referers=["https://localhost:8888"],
    expiration=three_days
)
new_token_1 = new_developer_credential.generate_token(slot=1)
```

```python
print(f"New Item created: \n\t{new_developer_credential} \n\tPartial key: {new_token_1["access_token"][-12:]}")
```

```python
portal.content.get(new_developer_credential._item.id)
```
