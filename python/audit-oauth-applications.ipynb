## DTS2026 Audit OAuth apps

```python
def make_link(val):
    return f'<a href="{val}" target="_blank">Link</a>'
```

#### import modules

```python
from arcgis.gis import GIS,pd
```

#### connect to portal

```python
portal = GIS("home")
```

#### create the search query

```python
# find OAuth apps of a specific user
q = f'owner: {portal.properties.user.username} AND type:Application AND typekeywords:Application AND typekeywords:"Registered App" AND orgid: {portal.properties.id}'
# find all apps in org
# q = f'type:Application AND typekeywords:Application AND typekeywords:"Registered App" AND orgid: {portal.properties.id}'
```

#### perform the search

```python
items = portal.content.advanced_search(query=q,sort_field="owner", sort_order="desc")
```

```python
print(f"{items['total']} OAuth applications found")
```

```python
# Get the details of each registered app
apps = [
    {
        "title": itm.title,
        "typekeywords": itm.typeKeywords,
        "owner": itm.owner,
        "itemId": (item_id := itm.app_info["itemId"]),
        "redirectUris": ", ".join(itm.app_info["redirect_uris"]),
        "link": f"https://www.arcgis.com/home/item.html?id={item_id}"
    }
    for itm in items['results']
]
```

#### Output the details of each app

```python
df = pd.DataFrame(apps)
df.style.format({"link":make_link})
```
